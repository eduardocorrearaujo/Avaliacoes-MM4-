# Importando os pacotes necessários

```
# importar as bibliotecas necessárias
import pandas as pd
import numpy as np
import plotly.graph_objects as go

```

# Importando os dados 

Os dados importados são da plataforma [Our World in Data](https://ourworldindata.org/coronavirus).

```
dadosCOVID = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

```
dadosge = dadosCOVID.loc[dadosCOVID.location == 'Germany']
dadosge.head(72)
```

## Plotando dados básicos sobre a pandemia disponíveis no Our World in Data

Dando um duplo clique na legenda é possível visualizar a curva selecionada.

```
fig  = go.Figure()

fig.add_trace(go.Scatter( x = dadosge.date, y = dadosge.new_cases, name = "Casos Diários"))
fig.add_trace(go.Scatter( x =  dadosge.date, y = dadosge.new_cases_smoothed, name = "Casos Diários - Média Móvel de 7 dias"))
fig.add_trace(go.Scatter( x = dadosge.date, y = dadosge.total_cases, name = "Casos Acumulados"))
fig.add_trace(go.Scatter( x =  dadosge.date, y = dadosge.new_deaths, name = "Óbitos Diários"))
fig.add_trace(go.Scatter( x =  dadosge.date, y = dadosge.new_deaths_smoothed, name = "Óbitos Diários - Média Móvel de 7 dias"))
fig.add_trace(go.Scatter( x =  dadosge.date, y = dadosge.total_deaths, name = "Óbitos Acumulados"))

fig.update_layout(
    xaxis_title = "Data",
    yaxis_title = "",
    plot_bgcolor = 'rgba(0,0,0,0)')

fig.update_xaxes( showgrid=True, gridwidth=1, gridcolor='rgb(211,211,211)',
        showline=True, linewidth=1, linecolor='black'
                 )


fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='rgb(211,211,211)',
        showline=True, linewidth=1, linecolor='black')

fig.update_layout(title = "Dados COVID-19 - Alemanha")
fig.show()
```
